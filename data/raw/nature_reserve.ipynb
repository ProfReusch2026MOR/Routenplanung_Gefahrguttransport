```python
import geopandas as gpd

url = "https://geodienste.bfn.de/ogc/wfs/schutzgebiet"

gdf = gpd.read_file(
    url,
    layer="Naturschutzgebiete",
    rows=slice(None)
)
gdf = gdf.to_crs(epsg=4326)
```

```python
gdf_filtered = gdf[
    gdf["BL"].isin(["NW", "NI"])
]
```

```python
gdf_filtered.to_csv("../processed/nature_reserves_filtered.csv",
                index=False,
                encoding="utf-8")
```
