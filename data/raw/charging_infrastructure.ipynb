```python
import pandas as pd

df = pd.read_csv(
    "Ladesaeulenregister_BNetzA_2026-04-22.csv",
    sep=";",
    encoding="cp1252",
    header=10,        
    low_memory=False
)
```

```python
cols = [
    "Ladeeinrichtungs-ID",
    "Betreiber",
    "Anzahl Ladepunkte",
    "Nennleistung Ladeeinrichtung [kW]",
    "Breitengrad",
    "Längengrad",
    "Ort",
    "Postleitzahl",
    "Bundesland"
]

df_clean = df[cols].copy()
```

```python
for col in [
    "Nennleistung Ladeeinrichtung [kW]",
    "Breitengrad",
    "Längengrad"
]:
    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .str.strip()
    )
```

```python
df_clean["Nennleistung Ladeeinrichtung [kW]"] = pd.to_numeric(
    df_clean["Nennleistung Ladeeinrichtung [kW]"],
    errors="coerce"
)

df_clean["Breitengrad"] = pd.to_numeric(
    df_clean["Breitengrad"],
    errors="coerce"
)

df_clean["Längengrad"] = pd.to_numeric(
    df_clean["Längengrad"],
    errors="coerce"
)
```

```python
df_clean = df_clean[
    df_clean["Nennleistung Ladeeinrichtung [kW]"] >= 150
]
```

```python
df_clean = df_clean.dropna(
    subset=["Breitengrad", "Längengrad"]
)
```

```python
df_clean.to_csv("../processed/charging_infrastructure_clean.csv",
                index=False,
                encoding="utf-8")
```
